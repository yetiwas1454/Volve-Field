```python
b=pd.read_excel("Volve production data.xlsx",sheet_name=0,usecols=['DATEPRD','NPD_WELL_BORE_NAME','AVG_DOWNHOLE_PRESSURE','AVG_CHOKE_SIZE_P','BORE_OIL_VOL','BORE_GAS_VOL','BORE_WAT_VOL','BORE_WI_VOL'])
b
```

```python
b.info()
```

```python
b[b["NPD_WELL_BORE_NAME"]=='15/9-F-1 C']
c=b[b["NPD_WELL_BORE_NAME"]=='15/9-F-1 C']
c
```

```python
x =c['BORE_OIL_VOL']
y =c['BORE_GAS_VOL']
z=y/x
c["GOR"]=z
c
```

```python
fig,ax1=plt.subplots()
ax1.plot(c["DATEPRD"],c["BORE_WAT_VOL"],color="blue")
ax1.set_xlabel("Months")
ax1.set_ylabel("Water production",color="blue")
ax1.tick_params(axis='y',labelcolor='blue')
ax2=ax1.twinx()
ax2.plot(c["DATEPRD"],c["AVG_CHOKE_SIZE_P"],color="brown")
ax2.set_ylabel("choke size (%)",color="brown")
ax2.tick_params(axis='y',labelcolor='brown')
plt.title('Water Production vs choke size')
fig.tight_layout()
```

```python
fig,ax1=plt.subplots()
ax1.plot(c["DATEPRD"],c["BORE_OIL_VOL"],color="green")
ax1.set_xlabel("Months")
ax1.set_ylabel("Oil production",color="green")
ax1.tick_params(axis='y',labelcolor='green')
ax2=ax1.twinx()
ax2.plot(c["DATEPRD"],c["GOR"],color="orange")
ax2.set_ylabel("choke size (%)")
ax2.tick_params(axis='y')
plt.title('Oil Production vs GOR')
fig.tight_layout()
```

```python
fig=plt.figure(figsize=(10,5))
plt.plot(c["DATEPRD"],c['BORE_OIL_VOL'],color="green")
plt.plot(c["DATEPRD"],c['BORE_WAT_VOL'],color="royalblue")
```
